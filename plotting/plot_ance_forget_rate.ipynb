```python
import os
import json
import seaborn
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib.pyplot import MultipleLocator
import brewer2mpl

plt.rc('font',family='Times New Roman')

if __name__ == "__main__":
    ## Color
    bmap = brewer2mpl.get_map('Set2', 'qualitative', 8)
    colors = bmap.mpl_colors
    
    plt.figure(figsize=(6.5,6.5))
    
    x = [i for i in range(4)]
    
    ## Vanilla BERT
    y1 = [27.812120356544234, 23.77922571126916, 29.490057442353844, 28.442813144337585]
    
    ## Vanilla coCondenser
    y2 = [24.705779428519165, 19.516879780649344, 24.09675924913359, 22.764192078959873]
    
    
    plt.rcParams.update({'font.size':30})
    ax=plt.gca()

    markersize=11
    linewidth=2.0

    ## x-刻度
    x_major_locator=MultipleLocator(1)
    ax.xaxis.set_major_locator(x_major_locator)

    plt.xlim(-0.1, 3.1)
    plt.ylim(12, 34)

    color_1 = colors[0]
    color_2 =colors[1]

    l1=plt.plot(x,y1, color=color_1, marker='s', linewidth=linewidth, markersize=markersize, label="ANCE (BERT)")
    l11=plt.plot(x,y2, color=color_2, marker='o', linewidth=linewidth, markersize=markersize+1, label="ANCE (coCondenser)")

    ax.set_xticks([-0.1, 1, 2, 3.1])
    ax.set_xticklabels(['Epi-2', 'Epi-3', 'Epi-4', 'Epi-5'])

    plt.ylabel('Forgetting rate (%)')
    # plt.legend()
    plt.legend(prop = {'size':25.5}, frameon=False, bbox_to_anchor=(1.0, 0.32))
    plt.grid(linestyle = ':', linewidth = 0.5)

    plt.show()
```
