```python
import os
import numpy as np
import matplotlib.pyplot as plt

import brewer2mpl
plt.rc('font',family='Times New Roman')
bmap = brewer2mpl.get_map('Set2', 'qualitative', 8)
colors = bmap.mpl_colors

if __name__ == "__main__":

    plt.figure(figsize=(6.5,6.5))
    plt.rcParams.update({'font.size':30})
    
    plt.ylim(0,155)
    width = 0.52

    alpha = 0.7

    iter_1 = [100, 0, 0, 0, 0]
    iter_2 = [6.535550498774892, 93.46444950122508, 0, 0, 0]
    iter_3 = [2.236829498330764, 67.57213023538979, 30.191040266279423, 0, 0]
    iter_4 = [1.4670806074125087, 74.68303217210355, 18.6213140809717, 5.228573139512255, 0]
    iter_5 = [2.133076086251379, 68.78115783642976, 22.10217899007441, 2.680358648841514, 4.303228438402923]

    labels = ["Epi-1/BM25-Neg", "Epi-2-Neg", "Epi-3-Neg", "Epi-4-Neg", "Epi-5-Neg"]
    
    data_list = [iter_1, iter_2, iter_3, iter_4, iter_5]

    x = range(len(data_list))
    
    bottom_y = [0] * len(data_list)
    for idx, data in enumerate(zip(iter_1, iter_2, iter_3, iter_4, iter_5)):
        y = data
        plt.bar(x, y, width, bottom=bottom_y, color=colors[idx], alpha=alpha, label=labels[idx])
        bottom_y = [(a+b) for a, b in zip(y, bottom_y)]
    
    names = ['Epi-1', 'Epi-2', 'Epi-3', 'Epi-4', 'Epi-5']
    plt.legend(loc=2, ncol=2, prop = {'size':22.1}, frameon=False, columnspacing=0.2, handletextpad=0.2, bbox_to_anchor=(-0.03, 1.0))

    plt.xticks([0,1,2,3,4], names)
    plt.yticks([0, 20, 40, 60, 80, 100])
    plt.ylabel('Negative Ratio (%)')

    plt.show()
```
